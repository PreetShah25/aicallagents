# ACE Pro Demo
Quick, offline demo using the LocalStub (no API keys needed).

```python
from ace_pro.agent_loop import run_agent_call
from ace_pro.tools.summarize import summarize_extractive
intro = "System: Expert call on production Voice AI (no MNPI)."
transcript = run_agent_call(intro, knowledge_index=None, turns=3)
bullets = summarize_extractive(transcript, top_k=5)
print(transcript[:500])
print('\nHighlights:')
for b in bullets: print('-', b)
```
